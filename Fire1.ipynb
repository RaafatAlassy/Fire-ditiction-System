```python
 # Make sure you have installed this library
from ultralytics import YOLO
import cv2
import math
import cvzone 
import numpy as np
import smtplib
import playsound
import threading
import vonage
```

```python
#Import model
model = YOLO('fire.pt')

# Update the video source as needed
cap = cv2.VideoCapture("fire4.mp4")  # Update the video source as needed

if not cap.isOpened():
    print("Failed to open the video capture.")
    exit()

# Update with your class names
classnames = ["fire"]  



Alarm_Status = False
# Fire_Reported = 0

def play_alarm_sound_function():
	while True:
		playsound.playsound('alarm.mp3',True)




while True:
    ret, frame = cap.read()

    if not ret:
        print("Failed to read frame from the video capture.")
        break

    frame = cv2.resize(frame, (1024,720))
    result = model(frame, stream=True)  

    for info in result:
        boxes = info.boxes
        for box in boxes:
            confidence = box.conf[0]
            confidence = math.ceil(confidence * 100)
            Class = int(box.cls[0])
            if confidence > 60:
                x1, y1, x2, y2 = box.xyxy[0]
                x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 255), 5)
                cvzone.putTextRect(frame, f'{classnames[Class]} {confidence}%', [x1 + 8, y1 + 100],scale=2, thickness=2)

                #Alarm
                if Alarm_Status == False:
                    threading.Thread(target=play_alarm_sound_function).start()
                    Alarm_Status = True

                    # If you want to check Alarm_Status again, use a separate if statement
                    if Alarm_Status == False:
                        threading.Thread(target=play_alarm_sound_function).start()
                        Alarm_Status = True
                
                
               

    cv2.imshow('frame', frame)
    if cv2.waitKey(1) == ord(' '):
        break

cap.release()
cv2.destroyAllWindows()
```
